#### PCA (Principal Component Analysis)?
* PCA is a mathematical technique used to reduce the number of features in your dataset while keeping as much information (variance) as possible.
#### Why Do We Need PCA?
- When your dataset has too many correlated features, it can:
* Slow down model training
* Cause overfitting
* Make visualization and interpretation hard 
- PCA helps by:
* Removing redundancy
* Reducing dimensions
* Keeping only the “essence” of the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
df = sns.load_dataset('tips')

X = df[['total_bill', 'tip', 'size']]


scalar = StandardScaler()
scaled_df = scalar.fit_transform(X)

pca = PCA(n_components=2)
fit_pca = pca.fit_transform(scaled_df)

plt.scatter(fit_pca[:,0], fit_pca[:,1], c=pd.Categorical(df['sex']).codes, cmap='rainbow')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
