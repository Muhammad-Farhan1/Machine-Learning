**Rolling statistics** are a set of calculations that `provide insights` into data trends over a specified moving window.\
They are *commonly used in time series analysis* to smooth out short-term fluctuations and highlight longer-term trends or cycles.\
Rolling statistics can include measures such as rolling mean, rolling median, rolling standard deviation, and rolling sum.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = yf.download('AAPL', start='2020-01-01', end='2025-01-01', auto_adjust=True)
df.head()
```

```python
df['Open 20 Day MA'] = df['Open'].rolling(window=20, min_periods=1).mean()
df['Open 50 Day MA'] = df['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df[['Open', 'Open 20 Day MA', 'Open 50 Day MA']].plot(figsize=(16,8), xlim=['2020-01-01', '2021-01-01'])
```

**Expanding** on the concept of moving averages, rolling statistics help analysts and data scientists to better understand the underlying patterns in their data by continuously updating the calculations as new data points become available.\

This technique is particularly `useful` in fields such as finance, economics, and environmental science, where data is collected over time and trends need to be identified amidst noise and volatility.

Difference between Moving Average and Rolling Statistics:
- **Moving Average**: A moving average is a specific type of rolling statistic that calculates the average of a subset of data points over a defined window. It is primarily used to smooth out short-term fluctuations and highlight longer-term trends in time series data. The window size can be adjusted to capture different levels of detail in the data.

- **Rolling Statistics**: Rolling statistics encompass a broader range of calculations that can be performed over a moving window, including not only the mean (moving average) but also other statistical measures such as median, standard deviation, variance, minimum, maximum, and sum. Rolling statistics provide a more comprehensive view of the data's behavior over time, allowing analysts to assess variability, trends, and other characteristics beyond just the average.

Difference between Rolling and Expanding:
- **Rolling**: In rolling calculations, a fixed-size window moves across the data set, calculating statistics for each position of the window. As new data points enter the window, old data points exit, maintaining a constant window size. This approach is useful for analyzing local trends and fluctuations within a specific timeframe.
- **Expanding**: Expanding calculations, on the other hand, start with a small window that grows as more data points are added. The window begins with the first data point and expands to include all subsequent data points, calculating statistics over the entire range of data up to the current point. This method is useful for understanding cumulative trends and overall behavior of the data set over time.

```python
df['Open_Expanding_Mean'] = df['Open'].expanding(min_periods=1).mean()
```

```python
df.tail()
```

```python
df[['Open_Expanding_Mean', 'Open']].plot(figsize=(12,6), xlim=['2020-01-01', '2025-01-01'])
```

**Exponential Moving Average (EMA**) is a type of moving average that gives more weight to recent data points, making it more responsive to recent price changes compared to the Simple Moving Average (SMA). The EMA is calculated using a smoothing factor, which determines the degree of weighting decrease for older data points. This characteristic makes the EMA particularly useful in financial markets for identifying trends and potential reversals more quickly than the SMA.

**Exponential weight moving average** is a type of moving average that places greater emphasis on more recent data points while still considering older data points, albeit with exponentially decreasing weights. This approach allows the average to be more responsive to recent changes in the data, making it particularly useful for time series analysis where recent trends are more relevant.

```python
# code for Exponential Moving Average (EMA)
df['Open_EMA_20'] = df['Open'].ewm(span=20, adjust=False).mean()
```

```python
df[['Open', 'Open_EMA_20']].plot(figsize=(12,6), xlim=['2020-01-01', '2021-01-01'])
```
